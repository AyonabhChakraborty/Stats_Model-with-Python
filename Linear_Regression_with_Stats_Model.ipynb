```python
l1 = ['battery','weight']
for i in l1:
  df[i] = df[i].fillna(statistics.mode(df[i]))
```

```python
l = ['normalized_used_price','weight']
for i in l:
  Q1 = df[i].quantile(0.25)
  Q3 = df[i].quantile(0.75)
  IQR = Q3 - Q1
  lf = Q1 - 1.5*(IQR)
  uf = Q3 + 1.5*(IQR)
  x = df[i].mean()
  df[i] = np.where((df[i]>uf)|(df[i]<lf),x,df[i])

```

```python
mu, sigma = statistics.mean(df1.screen_size), statistics.stdev(df1.screen_size)


desired_min = df1.screen_size.min()
desired_max = df1.screen_size.max()


transformed_screen_sizes = stats.truncnorm(
    (desired_min - mu) / sigma, (desired_max - mu) / sigma, loc=mu, scale=sigma).rvs(len(df1.screen_size))


df1['screen_size'] = transformed_screen_sizes


plt.hist(df1['screen_size'], density=True, bins=20, alpha=0.5, label='Transformed Data')
plt.xlabel('Screen Size')
plt.ylabel('Probability Density')
plt.title('Transformed Screen Size Distribution')
plt.legend()
plt.show()
```

# Model Building with StatsModel

```python
import statsmodels.api as sm
from sklearn.model_selection import train_test_split
```

```python
x = df1.drop('normalized_used_price',axis=1)
```

```python
y = df1.normalized_used_price
```

```python
x
```

```python
x = x.drop(['device_brand', 'os_Others'], axis=1)
```

```python
x_c = sm.add_constant(x)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_c,y,test_size=0.30,random_state=1)
```

```python
model = sm.OLS(y_train,x_train.astype(float)).fit()
```

```python
model.summary()
```

```python
y_predict = model.predict(x_test.astype(float))
```

```python
residual = y_test - y_predict
```

```python
np.sqrt((residual**2).mean())
```

```python
y_test.mean()
```

```python
np.sqrt((residual**2).mean())/y_test.mean()
```
